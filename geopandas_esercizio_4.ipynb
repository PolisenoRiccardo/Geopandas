```
# caricare il geodataframe delle fontanelle Fontanelle.zip
```

```
fontane = gpd.read_file('/content/drive/MyDrive/Colab Notebooks/Dataset/Fontanelle.zip')
fontane
```

```
# visualizzare il geodataframe su una mappa
```

```
fontane3857 = fontane.to_crs(3857)
ax = fontane3857.plot(figsize=(12, 6), markersize=10)
contextily.add_basemap(ax)
```

```
# visualizzare su una mappa le fontanelle e i quartieri in cui si trovano.
# non visualizzare i quartieri in cui non ci sono fontanelle
```

```
fontanelle = gpd.GeoSeries(cascaded_union(fontane3857.geometry), crs=3857)
ax = quartieri3857[quartieri3857.intersects(fontanelle.geometry.item())].plot(figsize=(20, 10), markersize=10, edgecolor='k', facecolor='None')
fontanelle.plot(ax = ax)
contextily.add_basemap(ax)
```

```
# colorare i quartieri in modo proporzionale al numero di fontanelle presenti (giallo molte fontanelle, blu scuro poche fontanelle)
# non evidenziare i quartieri senza fontanelle
```

```
quartiericonfontane
```

```
joined = gpd.sjoin(fontane3857, quartieri3857, predicate='intersects', how='left')
fontaneinquartieri = joined.groupby('NIL', as_index=False).count()[['NIL','Licenza']]
quartiericonfontane = quartieri3857.merge(fontaneinquartieri, on='NIL')
ax = quartiericonfontane.plot(figsize=(12, 6),  column='Licenza', legend=True, cmap='viridis', edgecolor='k')
contextily.add_basemap(ax)
```

```
# visalizzare la mappa precedente facendo in modo che i quartieri con fontanelle siano colorati con gradazioni di rosso
# e quelli senza fontanelle con gradazioni di grigio
```
